```python
median = df_host['host_acceptance_rate'].median()
df_host['host_acceptance_rate'] = df_host['host_acceptance_rate'].fillna(median)
```

```python
median = df_host['host_response_rate'].median()
df_host['host_response_rate'] = df_host['host_response_rate'].fillna(median)
```

```python
df_host.head()
```

```python
df_host.shape
```

```python
df_host.isna().sum()
```

```python
df_host = df_host.drop(columns = ['host_neighbourhood', 'host_name','host_location'])
```

```python
df_host.dropna(inplace = True)
```

```python
df_host.shape
```

```python
df_host = df_host.rename(columns={'location': 'host_location'})
```

```python
df_host.columns
```

```python
df_host.dtypes
```

```python
df_host['host_since'] = pd.to_datetime(df_host['host_since'])
```

```python
# Check for duplicates in 'host_id'
duplicates = df_host.duplicated(subset='host_id', keep=False)

# Drop duplicate rows
df_host = df_host[~duplicates]

# Check the result
print(df_host['host_id'].duplicated().sum())
```

```python
df_reviews_t = pd.read_csv("/content/drive/MyDrive/Canada/Toronto/review_toronto_df.csv")
df_reviews_t.head()
```

```python
df_reviews_v = pd.read_csv("/content/drive/MyDrive/Canada/Vancouver/review_vancouver_df.csv")
df_reviews_v.head()
```

```python
# Check if the two dataframes have the same columns
if df_reviews_t.columns.equals(df_reviews_v.columns):
    print("The two dataframes have the same columns.")
else:
    print("The two dataframes do not have the same columns.")
```

```python
df_reviews = pd.merge(df_reviews_t, df_reviews_v, how='outer')
df_reviews.head()
```

```python
df_reviews['comments'] = df_reviews['comments'].str.replace('<br/>', '')
```

```python
df_reviews['id'] = df_reviews['id'].astype(str)
df_reviews = df_reviews[df_reviews['id'].str.isnumeric()]
```

```python
df_reviews['id'] = df_reviews['id'].astype(int)
```

```python
for i in df_reviews.columns:
    print(df_reviews[i].dtypes)
    print(df_reviews[i].isna().sum())
    print('>>>>>>>>>>>>>>>>')
    print(df_reviews[i].value_counts())
    print('________________________')
```
